# Functions

```python
def euclidian_norm(vect_1):
    
    return scalar_multiplication(vect_1,vect_1)**0.5 


def scalar_multiplication(vect_1,vect_2):
    
    if type(vect_1) != list or type(vect_2) != list:
        raise ValueError('Not lists')
    
    if len(vect_1) != len(vect_2):
        raise ValueError('Unequal dimensions')
        
    result = 0
    for i in range(len(vect_1)):
        result += vect_1[i]*vect_2[i]
        
    return result
```

```python
scalar_multiplication(1,[1,3])
```

```python
euclidian_norm([4,3])
```

# Class

```python
class Vector:
    
    def __init__(self,vect):
        
        self.vect = vect
        
        
    def get_norm(self):
        self.norm = self.multiply_by(self.vect)**0.5
        
        return self.norm
        
        
    def multiply_by(self,vect_2):
        
        result = 0
        for i in range(len(self.vect)):
            result += vect_2[i]*self.vect[i]
            
        return result
```

```python
obj = Vector([3,4])

obj.get_norm()
```

# Lambdas, Sum, Map, Filter, Reduce, List Generator

```python
lambda_scalar = lambda vect_1 , vect_2 : sum(vect_1[i]*vect_2[i] for i in range(len(vect_1)))

lambda_euclidian_norm = lambda vect : lambda_scalar(vect,vect)**0.5
```

```python
lambda_euclidian_norm([4,3])
```

```python
map( lambda x: x**2,[1,2,3] ) 
```

```python
list(map( lambda x: x**2,[1,2,3] )) 
```

```python
list(filter( lambda x: x%2 == 0,[1,2,3,4] ))
```
